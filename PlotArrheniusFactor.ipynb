```python
# Calculate the points for temperature and ice softness

tau0 = 0.5

def Arr_n3(A0, T):
    Q = 60e3
    R = 8.314
    return A0 * np.exp(-Q/(R*T))

def Arr_n4(A0, T):
    Q = 60e3
    R = 8.314
    return A0 * np.exp(-Q/(R*T)) / tau0

A0_n3 = 1.258e13
A0_n4 = 5.032e13

Temp = [273, 263]

Arr_n3_lst = [Arr_n3(A0_n3, 273), Arr_n3(A0_n3, 263)]
Arr_n4_lst = [Arr_n4(A0_n4, 273)*tau0, Arr_n4(A0_n4, 263)*tau0]
Arr_n4_lst_tau3 = [Arr_n4(A0_n4, 273)*3, Arr_n4(A0_n4, 263)*3]
Arr_n4_lst_tau1 = [Arr_n4(A0_n4, 273)*1, Arr_n4(A0_n4, 263)*1]
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 8))
ax.plot(Temp, Arr_n3_lst, color='black')
ax.plot(Temp, Arr_n4_lst, color='red')
ax.plot(Temp, Arr_n4_lst_tau1, color='green')
```
